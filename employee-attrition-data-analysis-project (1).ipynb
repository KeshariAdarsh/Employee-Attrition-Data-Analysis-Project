# **Attrition Count**

# **Importing Libraries For Graphical/Pictorial Representation**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Pie Chart Representation of Attrition Count**

```python
plt.pie(attrition_count['Attrition'] , labels = ['No' , 'Yes'] , explode = (0.2,0))
```

# **Bar Graph Representation of Attrition Count**

```python
sns.countplot(df['Attrition'])
```

# **Dropping Columns of No Use**

```python
df.drop(['EmployeeCount', 'EmployeeNumber'],axis=1)
```

```python
df
```

# **Making Columns Of Attrition "Yes" and "No"**

```python
attrition_dummies=pd.get_dummies(df['Attrition'])
attrition_dummies
```

# **Concatinating Coumns of Attrition in the dataset**

```python
df=pd.concat([df,attrition_dummies],axis=1)
df.head()
```

# **Dropping Column Of Attrition "No"**

```python
df.drop(['Attrition','No'],axis=1)
```

# **Analysis Showing Gender Wise Employee Attrition**

```python
sns.barplot(x='Gender' ,y='Yes',data=df)
```

# **Analysis Showing Department Wise Employee Attrition**

```python
sns.barplot(x = 'Department', y = 'Yes', data = df)
```
